# 1️ Time Feature Analysis  

##  Important Theory  

###  What is the Time Feature?

In the dataset, **Time** represents the number of seconds elapsed since the first transaction in the dataset.

It does NOT represent actual clock time directly.  
Therefore, we need to convert it into meaningful time-based features.


##  Why Convert Time?

Fraudulent transactions often follow **time-based patterns**:

- Late-night frauds are common.
- Certain hours may show higher fraud rates.
- Fraudsters may target low-activity hours.

To analyze this properly, we convert the Time feature into:

### 1️ Hour_of_day
Converts seconds into:
- 0 to 23 hour format
- Helps identify peak fraud hours

Formula:
Hour_of_day = (Time / 3600) % 24



### 2️ Day Segment
Categorizing hours into broader segments:

| Hour Range | Segment |
|------------|----------|
| 5 – 11     | Morning  |
| 12 – 17    | Afternoon |
| 18 – 21    | Evening  |
| 22 – 4     | Night    |

This helps identify:
- Whether fraud is more common at night
- Behavioral fraud patterns during specific time blocks


````python
# Convert Time to Hour of Day
df['Hour_of_day'] = (df['Time'] // 3600) % 24

# Create Day Segment
def get_day_segment(hour):
    if 6 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 18:
        return "Afternoon"
    else:
        return "Night"

df['Day_segment'] = df['Hour_of_day'].apply(get_day_segment)

# Transaction distribution over time
plt.figure()
df['Hour_of_day'].value_counts().sort_index().plot(kind='bar')
plt.title("Transaction Distribution by Hour")
plt.xlabel("Hour")
plt.ylabel("Count")
plt.show()

# Fraud rate by hour
fraud_rate = df.groupby('Hour_of_day')['Class'].mean()

plt.figure()
fraud_rate.plot(kind='bar')
plt.title("Fraud Rate by Hour")
plt.xlabel("Hour")
plt.ylabel("Fraud Rate")
plt.show()

# Peak fraud hour
peak_hour = fraud_rate.idxmax()
print("Peak Fraud Hour:", peak_hour)
````

# 3️ Feature vs Target Analysis  

##  Important Theory  

### 🔹 What is Feature vs Target Analysis?

In fraud detection:

- **Features (Independent Variables)** → V1, V2, V3 … V28, Amount, Time  
- **Target (Dependent Variable)** → Class  
  - 0 = Non-Fraud  
  - 1 = Fraud  

Feature vs Target analysis helps us understand:

- Which features influence fraud the most  
- Which features contribute positively or negatively  
- Which features are strong fraud indicators  

---

##  What We Analyze

### 1️ Mean Difference of Each V-Feature

We compare:

- Mean of each feature when Class = 0  
- Mean of each feature when Class = 1  

Why?

If a feature shows a large mean difference between fraud and non-fraud:

➡ It may be a strong fraud predictor  

Example:

If:
Mean(V14 | Fraud) is very different from Mean(V14 | Non-Fraud)

Then:
V14 likely contributes significantly to fraud detection.


### 2️ Correlation with Target (Class)

We compute:

Correlation(feature, Class)

Interpretation:

| Correlation Value | Meaning |
|-------------------|----------|
| Close to +1       | Strong positive relation with fraud |
| Close to -1       | Strong negative relation with fraud |
| Close to 0        | Weak or no relation |

---

##  What Correlation Tells Us

###  Positive Correlation

If correlation > 0:

As feature value increases → Fraud probability increases.

These are **positive contributors**.

---

###  Negative Correlation

If correlation < 0:

As feature value increases → Fraud probability decreases.

These are **negative contributors**.

````python
# Mean difference
mean_diff = df.groupby('Class').mean(numeric_only=True).T
mean_diff['Difference'] = mean_diff[1] - mean_diff[0]
print(mean_diff[['Difference']].sort_values(by='Difference', ascending=False).head())

# Correlation with Class
corr = df.corr(numeric_only=True)['Class'].sort_values(ascending=False)
print(corr.head(10))

# Top correlated features visualization
top_features = corr[1:6].index

for feature in top_features:
    plt.figure()
    sns.boxplot(x='Class', y=feature, data=df)
    plt.title(f"{feature} vs Class")
    plt.show()
````

# 4️ Fraud vs Non-Fraud Comparison  

## Important Theory  

###  What is Fraud vs Non-Fraud Comparison?

In fraud detection, we compare statistical properties of each feature between:

- **Class = 0 → Non-Fraud**
- **Class = 1 → Fraud**

This helps us understand how fraudulent transactions behave differently from normal transactions

## For Each Feature, We Calculate

For every feature (V1–V28, Amount, Time), compute:

### 1️ Mean (Average)

Formula:
Mean = Sum of values / Total observations

Purpose:
- Identifies overall trend of the feature
- Large difference between Fraud and Non-Fraud means → Strong fraud signal

Example:
If Mean(V14 | Fraud) is very different from Mean(V14 | Non-Fraud),
then V14 may be an important fraud indicator.



### 2️ Median

Median = Middle value when data is sorted

Why important?

- Less sensitive to outliers
- Better measure when data is skewed
- Fraud datasets are usually highly skewed

If:
Mean ≠ Median → Data may contain outliers or skewness

### 3️ Standard Deviation (Std)

Std measures spread of data.

Formula:
Std = √(Variance)

Purpose:
- Shows variability of feature values
- High Std in fraud class → Unstable or extreme fraud behavior
- Low Std → More consistent behavior

````python
numeric_cols = df.select_dtypes(include=[np.number]).columns
group_stats = df.groupby('Class')[numeric_cols].agg(['mean','median','std'])

print(group_stats)
````

# 1️ Correlation-Based Feature Reduction


###  What is Correlation?

Correlation measures the **linear relationship** between two features (variables).

It tells us how strongly two features move together.


# 2️ PCA (Principal Component Analysis)

###  What is PCA?

PCA (Principal Component Analysis) is a **dimensionality reduction technique** that transforms original features into a new set of features called:

> **Principal Components**

These components are:
- Linear combinations of original features
- Uncorrelated with each other
- Ordered by importance (variance explained)


###  Why PCA is Used?

PCA helps to:

-  Reduce number of features
-  Remove noise
-  Remove multicollinearity
-  Improve model performance
-  Speed up training
-  Avoid overfitting


### 🔹 How PCA Works (Conceptually)

1. Standardize the data  
2. Compute covariance matrix  
3. Calculate eigenvalues & eigenvectors  
4. Sort components by explained variance  
5. Select top N components  

````python
# 2️⃣ Separate Input and Target
X = df.drop("Class", axis=1)   # Input features only
y = df["Class"]                # Target (not used in reduction)

# Select only numeric columns for correlation analysis
X_numeric = X.select_dtypes(include=[np.number])

# 3️⃣ Remove Highly Correlated Features (>0.9)
corr_matrix = X_numeric.corr().abs()

upper = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)

to_drop = [column for column in upper.columns if any(upper[column] > 0.9)]

X_reduced_corr = X_numeric.drop(columns=to_drop)

print("Original Shape:", X_numeric.shape)
print("After Correlation Removal:", X_reduced_corr.shape)

# 4️⃣ Scale Features (Important before PCA)
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_reduced_corr)

# 5️⃣ Apply PCA (Keep 95% Variance)
pca = PCA(n_components=0.95)
X_final = pca.fit_transform(X_scaled)

print("Final Shape After PCA:", X_final.shape)

# 6️⃣ Plot Explained Variance Curve
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA Explained Variance")
plt.show()
````
