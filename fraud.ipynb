# 1️ Time Feature Analysis  

##  Important Theory  

###  What is the Time Feature?

In the dataset, **Time** represents the number of seconds elapsed since the first transaction in the dataset.

It does NOT represent actual clock time directly.  
Therefore, we need to convert it into meaningful time-based features.


##  Why Convert Time?

Fraudulent transactions often follow **time-based patterns**:

- Late-night frauds are common.
- Certain hours may show higher fraud rates.
- Fraudsters may target low-activity hours.

To analyze this properly, we convert the Time feature into:

### 1️ Hour_of_day
Converts seconds into:
- 0 to 23 hour format
- Helps identify peak fraud hours

Formula:
Hour_of_day = (Time / 3600) % 24



### 2️ Day Segment
Categorizing hours into broader segments:

| Hour Range | Segment |
|------------|----------|
| 5 – 11     | Morning  |
| 12 – 17    | Afternoon |
| 18 – 21    | Evening  |
| 22 – 4     | Night    |

This helps identify:
- Whether fraud is more common at night
- Behavioral fraud patterns during specific time blocks


````python
# Convert Time to Hour of Day
df['Hour_of_day'] = (df['Time'] // 3600) % 24

# Create Day Segment
def get_day_segment(hour):
    if 6 <= hour < 12:
        return "Morning"
    elif 12 <= hour < 18:
        return "Afternoon"
    else:
        return "Night"

df['Day_segment'] = df['Hour_of_day'].apply(get_day_segment)

# Transaction distribution over time
plt.figure()
df['Hour_of_day'].value_counts().sort_index().plot(kind='bar')
plt.title("Transaction Distribution by Hour")
plt.xlabel("Hour")
plt.ylabel("Count")
plt.show()

# Fraud rate by hour
fraud_rate = df.groupby('Hour_of_day')['Class'].mean()

plt.figure()
fraud_rate.plot(kind='bar')
plt.title("Fraud Rate by Hour")
plt.xlabel("Hour")
plt.ylabel("Fraud Rate")
plt.show()

# Peak fraud hour
peak_hour = fraud_rate.idxmax()
print("Peak Fraud Hour:", peak_hour)
````

# 3️ Feature vs Target Analysis  

##  Important Theory  

###  What is Feature vs Target Analysis?

In fraud detection:

- **Features (Independent Variables)** → V1, V2, V3 … V28, Amount, Time  
- **Target (Dependent Variable)** → Class  
  - 0 = Non-Fraud  
  - 1 = Fraud  

Feature vs Target analysis helps us understand:

- Which features influence fraud the most  
- Which features contribute positively or negatively  
- Which features are strong fraud indicators  



###  What We Analyze

### 1️ Mean Difference of Each V-Feature

We compare:

- Mean of each feature when Class = 0  
- Mean of each feature when Class = 1  

Why?

If a feature shows a large mean difference between fraud and non-fraud:

➡ It may be a strong fraud predictor  

Example:

If:
Mean(V14 | Fraud) is very different from Mean(V14 | Non-Fraud)

Then:
V14 likely contributes significantly to fraud detection.


### 2️ Correlation with Target (Class)

We compute:

Correlation(feature, Class)

Interpretation:

| Correlation Value | Meaning |
|-------------------|----------|
| Close to +1       | Strong positive relation with fraud |
| Close to -1       | Strong negative relation with fraud |
| Close to 0        | Weak or no relation |

###  What Correlation Tells Us

###  Positive Correlation

If correlation > 0:

As feature value increases → Fraud probability increases.

These are **positive contributors**.

###  Negative Correlation

If correlation < 0:

As feature value increases → Fraud probability decreases.

These are **negative contributors**.

````python
# Mean difference
mean_diff = df.groupby('Class').mean(numeric_only=True).T
mean_diff['Difference'] = mean_diff[1] - mean_diff[0]
print(mean_diff[['Difference']].sort_values(by='Difference', ascending=False).head())

# Correlation with Class
corr = df.corr(numeric_only=True)['Class'].sort_values(ascending=False)
print(corr.head(10))

# Top correlated features visualization
top_features = corr[1:6].index

for feature in top_features:
    plt.figure()
    sns.boxplot(x='Class', y=feature, data=df)
    plt.title(f"{feature} vs Class")
    plt.show()
````

# h) Correlation-Based Feature Reduction


###  What is Correlation?

Correlation measures the **linear relationship** between two features (variables).

It tells us how strongly two features move together.


### 2 PCA (Principal Component Analysis)

###  What is PCA?

PCA (Principal Component Analysis) is a **dimensionality reduction technique** that transforms original features into a new set of features called:

> **Principal Components**

These components are:
- Linear combinations of original features
- Uncorrelated with each other
- Ordered by importance (variance explained)


###  Why PCA is Used?

PCA helps to:

-  Reduce number of features
-  Remove noise
-  Remove multicollinearity
-  Improve model performance
-  Speed up training
-  Avoid overfitting


````python
# For correlation features reduction

# Separate features and target
target_col = "target" if "target" in df.columns else "Class"
X = df.drop(columns=[target_col])  
y = df[target_col]                

# Drop non-numeric columns before correlation
X = X.select_dtypes(include=[np.number])

# Compute correlation matrix
corr_matrix = X.corr().abs()

# Select upper triangle matrix
upper = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)

# Find features with correlation greater than threshold
threshold = 0.9

to_drop = [column for column in upper.columns if any(upper[column] > threshold)]

print("Highly correlated features to remove:", to_drop)

# Drop the correlated features
X_reduced = X.drop(columns=to_drop)

print("Original shape:", X.shape)
print("Reduced shape:", X_reduced.shape)
````

````python

# Separate features and target
X = df.drop(columns=[target_col])  
y = df[target_col]

# Keep only numeric columns for scaling/PCA
X = X.select_dtypes(include=[np.number])

# Step 1: Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 2: Apply PCA
pca = PCA(n_components=0.95)   # keep 95% variance
X_pca = pca.fit_transform(X_scaled)

print("Original shape:", X.shape)
print("Reduced shape:", X_pca.shape)

# Explained variance
print("Explained variance ratio:", pca.explained_variance_ratio_)
````

### Feature Reduction Results

After applying **Correlation-based Feature Removal** and **Principal Component Analysis (PCA)**, the dataset dimensions were reduced while retaining most of the important information.

### Interpretation

# Normality Test by Analyzing Skew (Before Scaling or Normalization)

Before applying any scaling or normalization techniques, it is important to check whether the numerical features in the dataset follow a normal (Gaussian) distribution. One common way to evaluate this is by analyzing the **skewness** of each feature.

**Skewness** measures the asymmetry of the distribution of data.

- If the skewness value is **close to 0**, the distribution is approximately **normal or symmetric**.
- A **positive skew (> 0)** indicates that the distribution is **right-skewed**, meaning the tail extends toward **higher values**.
- A **negative skew (< 0)** indicates that the distribution is **left-skewed**, meaning the tail extends toward **lower values**.

By calculating the skewness of each numerical feature **before scaling or normalization**, we can identify which variables are significantly skewed. Features with high skewness may require transformation techniques such as **log transformation, square root transformation, or Box-Cox transformation** to make the data more normally distributed.

Performing this step before scaling ensures that the chosen preprocessing technique (such as **StandardScaler** or **MinMaxScaler**) works more effectively and improves the overall performance of machine learning models.

````python
# Select only numeric columns
numeric_features = df.select_dtypes(include=[np.number])

# Calculate skewness
skew_values = numeric_features.skew()

# Convert to DataFrame for better visualization
skew_df = pd.DataFrame({
    "Feature": skew_values.index,
    "Skewness": skew_values.values
})

# Sort by skewness
skew_df = skew_df.sort_values(by="Skewness", ascending=False)

print("Skewness of Numerical Features (Before Scaling/Normalization)")
print(skew_df)
````

````python
# Threshold commonly used = 1 or -1
high_skew = skew_values[abs(skew_values) > 1]

print("\nHighly Skewed Features:")
print(high_skew)
````

# Scaling / Normalization (Numeric Features Only)

Scaling or normalization is a preprocessing step used to transform numeric features so that they are on a similar scale. Many machine learning algorithms, such as Logistic Regression, SVM, and KNN, perform better when features have comparable ranges. Scaling helps prevent features with larger values from dominating the learning process and improves model convergence and stability. Therefore, scaling or normalization is applied only to numeric features in the dataset.

````python
# Select only numeric features
numeric_features = df.select_dtypes(include=['int64', 'float64'])

# Initialize scaler
scaler = StandardScaler()

# Fit and transform the numeric data
scaled_data = scaler.fit_transform(numeric_features)

# Convert scaled data back to DataFrame
scaled_df = pd.DataFrame(scaled_data, columns=numeric_features.columns)

# Save the scaler weights for demo purposes
joblib.dump(scaler, "scaler_weights.pkl")

print("Scaling completed and scaler weights saved successfully.")
````

#   Normality Test by Analyzing Skew (After Scaling or Normalization)

After applying scaling or normalization, the distribution of numerical features is checked again using skewness to determine whether the data follows a normal distribution.

i. If some features are highly skewed, apply a **log1p transformation** (log(1 + x)) to those features before performing scaling or normalization again. This transformation helps reduce skewness and makes the distribution closer to normal.

````python
# Separate features
X = df.drop("Class", axis=1)

# Select only numeric columns
numeric_features = X.select_dtypes(include=['int64','float64'])

# Scale numeric features
scaler = StandardScaler()
X_scaled = pd.DataFrame(
    scaler.fit_transform(numeric_features),
    columns=numeric_features.columns
)

# Check skewness
skew_values = X_scaled.skew()

# Find highly skewed features
high_skew_features = skew_values[abs(skew_values) > 1].index

# Apply log1p transformation
for col in high_skew_features:
    numeric_features[col] = np.log1p(numeric_features[col])

# Scale again after transformation
X_scaled_final = pd.DataFrame(
    scaler.fit_transform(numeric_features),
    columns=numeric_features.columns
)

print(X_scaled_final.head())
````

# Model Training

Model training is the process of teaching a machine learning algorithm to learn patterns from the training dataset. The model uses the input features to predict the target variable. During training, the algorithm adjusts its internal parameters so that it can make accurate predictions on unseen data.

````python
X = df.drop("Class", axis=1)
numeric_features = X.select_dtypes(include=['int64','float64'])

scaler = StandardScaler()
X_final = pd.DataFrame(
    scaler.fit_transform(numeric_features),
    columns=numeric_features.columns
)

y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(
    X_final, y, test_size=0.2, random_state=42, stratify=y
)

sm = SMOTE(random_state=42)
X_train_balanced, y_train_balanced = sm.fit_resample(X_train, y_train)

model_lr = LogisticRegression(random_state=42, max_iter=1000)
model_lr.fit(X_train_balanced, y_train_balanced)

y_pred = model_lr.predict(X_test)
y_prob = model_lr.predict_proba(X_test)[:, 1]

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nROC-AUC Score:", roc_auc_score(y_test, y_prob))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:\n", cm)

plt.figure()
plt.imshow(cm)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")

for i in range(len(cm)):
    for j in range(len(cm)):
        plt.text(j, i, cm[i][j], ha='center', va='center')

plt.colorbar()
plt.show()
````
